## Preprocessing steps

```python
# dummy variables
months=pd.get_dummies(df_bike.months,drop_first=True,dtype=int)
weekdays=pd.get_dummies(df_bike.weekday,drop_first=True,dtype=int)
weathersit=pd.get_dummies(df_bike.weathersit,drop_first=False,dtype=int)
seasons=pd.get_dummies(df_bike.season,drop_first=True,dtype=int)
```

```python
# joining the dataframe
df_bike=pd.concat([months,weekdays,weathersit,seasons,df_bike],axis=1)
df_bike.head()
```

```python
df_bike.drop(['months', 'weekday', 'weathersit', 'season'], axis = 1, inplace = True)
```

## Train Test Split

```python
X=df_bike.drop(["count"],axis=1)
y=df_bike["count"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
X_train.shape
```

```python
# Rescaling
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
num_var = ['temp','humidity','windspeed']

X_train[num_var] = scaler.fit_transform(X_train[num_var])
X_test[num_var] = scaler.fit_transform(X_test[num_var])
```

```python
X_train.head()
```

## Feature Selection

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

estimator=LinearRegression()
selector=RFE(estimator,n_features_to_select=15)

selector=selector.fit(X_train,y_train)
selector.support_
```

```python
selected_features=X_train.columns[selector.support_]
```

```python
X_train=X_train[selected_features]
X_test=X_test[selected_features]
```

## Model Building

```python
import statsmodels.api as sm
X_train_sm=sm.add_constant(X_train) #X_train_sm["const"]=1
X_test_sm=sm.add_constant(X_test)
```

```python
model1=sm.OLS(y_train,X_train_sm)
res1=model1.fit()
res1.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train.columns

vif_data["VIF"]=[variance_inflation_factor(X_train.values,i) \
                 for i in range(len(X_train.columns))]
vif_data
```

```python
X_train_sm=X_train_sm.drop(["Clear"],axis=1)
X_test_sm=X_test_sm.drop(["Clear"],axis=1)
```

```python
model2=sm.OLS(y_train,X_train_sm)
res2=model2.fit()
res2.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) for i in range(len(X_train_sm.columns))]
vif_data
```

```python
X_train_sm=X_train_sm.drop(["Spring"],axis=1)
X_test_sm=X_test_sm.drop(["Spring"],axis=1)
```

```python
model3=sm.OLS(y_train,X_train_sm)
res3=model3.fit()
res3.summary()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) for i in range(len(X_train_sm.columns))]
vif_data
```

```python
res3.params
```

```python
y_pred = res3.predict(X_test_sm)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
